```python
import spacy
import el_core_news_md
nlp = el_core_news_md.load()
```

```python
import pandas as pd
df_texts = pd.read_csv('../data/votana_total_texts_annotated.csv', index_col=None, converters={'entities': eval})
```

```python
import networkx as nx
doc = nlp(df_texts.iloc[0]['text'])
# Load spacy's dependency tree into a networkx graph
edges = []
prev_sent_root = None
for token in doc:
    print(token.text,token.dep_,token.head.text)
    if token.dep_ == 'ROOT':
        if prev_sent_root != None:
            edges.append(('{0}'.format(prev_sent_root.idx),'{0}'.format(token.idx)))
        prev_sent_root = token
    for child in token.children:
        edges.append(('{0}'.format(token.idx),'{0}'.format(child.idx)))

graph = nx.Graph(edges)
#print(nx.is_connected(graph))
#nx.draw(graph, with_labels = True)
cnt = 0
for (start,end,entity1) in df_texts.iloc[0]['entities']:
    paths = []
    for (start2,end2,entity2) in df_texts.iloc[0]['entities']:
        if entity1 == 'ORGAN' and entity2 == 'PLANT':
            paths += [(nx.shortest_path_length(graph, source=str(start), target=str(start2)), 
                       nx.shortest_path(graph, source=str(start), target=str(start2)))]
    if paths != []:
        cnt += 1
        print(min(paths, key = lambda t: t[1]))
print(cnt)
```
